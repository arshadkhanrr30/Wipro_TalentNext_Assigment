Q1 Write a program to count the number of upper and lower case letters in a String.

```python
str = input("Enter a string: ")

upper = 0
lower = 0

for ch in str:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1

print("Uppercase letters ==>", upper)
print("Lowercase letters ==>", lower)
```

Q2 Write a program that will check whether a given String is Palindrome or not.

```python
str = input("Enter a string: ")

if str == str[::-1]:
    print("str is Palindrome")
else:
    print("str is Not a Palindrome")
```

Q3 Given string, return a new string made of n copies of the first 2 chars of the original string where n is the length of the string. The string length will be >=2. 
If input is "Wipro" then output should be "wiwiwiwiwi"

```python
str = input("Enter a string: ")

result = str[:2].lower() * len(str)

print(result)
```

Q4 Given a string, if the first or last character is 'x', return the string without those "x" character, else return the string unchanged. 
If the input is "xhix", then output is "Hi".

```python
s = input("Enter a string")

if s.startswith("x"):
    s = s[1:]

if s.endswith("x"):
    s = s[:-1]

print(s)
```

Q5 Given string and an integer n, return. string made of n repetitions of the last characters of the string. You may assume that n is between & and the length of the string inclusive. 
For example If the inputs are "Wipro" and 3, then the output should be "propropro".

```python
s = input("Enter a string")
n = int(input("Enter n: "))

result = s[-n:] * n

print(result)
```
