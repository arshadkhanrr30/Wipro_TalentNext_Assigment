Q1 Write a program to create a list of 5 integers and display the list items. Access individual elements through index.

```python
numbers = [10, 20, 30, 40, 50]

print("List", numbers)

print("First element ", numbers[0])
print("Second element ", numbers[1])
print("Third element ", numbers[2])
print("Fourth element ", numbers[3])
print("Fifth element ", numbers[4])
```

Q2 Write program to append a new item to the end of the list.

```python
numbers = [10, 20, 30, 40, 50]

new_item = int(input("Enter a new item to add to the list"))

numbers.append(new_item)

print("Updated List:", numbers)
```

Q3 Write a program to reverse the order of the items in the list.

```python
numbers = [10, 20, 30, 40, 50]

numbers.reverse()

print("Reversed List:", numbers)
```

Q4 Write a program to print the number of occurrences of a specified element in a list.

```python
numbers = [10, 20, 30, 20, 40, 20, 50]

element = int(input("Enter the element to count occurrences"))

count = numbers.count(element)

print("Number of occurrences=>", count)
```

Q5 Write a program to append the items of list1 to the front of list2.

Q6 Write a program to insert a new item before the second element in an existing list.

```python
numbers = [10, 20, 30, 40, 50]

new_item = int(input("Enter a new item"))

numbers.insert(1, new_item)

print("Updated List:", numbers)
```

Q7 Write a program to remove the item from a specified index in a list.

```python
numbers = [10, 20, 30, 40, 50]

index = int(input("Enter the index to remove: "))

numbers.pop(index)

print("Updated List:", numbers)
```
