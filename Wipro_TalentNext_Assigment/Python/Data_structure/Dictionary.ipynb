Q1 Write a program to add a key and value to a dictionary.
Sample Dictionary (0: 10, 1: 20) 
Expected Result: (0: 10, 1: 20, 2: 30)

```python
dict = {0: 10, 1: 20}

dict[2] = 30
dict[3] = 40
dict[4] = dict[2] + dict[3]

print(dict)
```

Q2 Write a program to concatenate the following dictionaries to create a new one. 
Sample Dictionary: dic1(1:10, 2:20) dic2=(3:30, 4:40) dic3=(5:50,6:60) 
Expected Result (1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60)

```python
dict1 = {1: 10, 2: 20}
dict2 = {3: 30, 4: 40}
dict3 = {5: 50, 6: 60}

dict = {}

dict.update(dict1)
dict.update(dict2)
dict.update(dict3)

print(dict)
```

Q3 Write a program to check if a given key already exists in a dictionary.

```python
dict = {
    1: "Apple",
    2: "Banana",
    3: "Mango"
}

key = int(input("Enter the key"))

if key in dict:
    print("Key exists")
else:
    print("Key does not exist")
```

Q4 Write a program to iterate over a dictionary using a for loop and print the keys alone, values alone, and both keys and values.

```python
dict = {
    1: "Apple",
    2: "Banana",
    3: "Mango"
}

print("Keys==>")
for key in dict:
    print(key, end=" ")

print("\nValues==>")
for value in dict.values():
    print(value)

print("Keys and Values==>")
for key, value in dict.items():
    print(key, ":", value)
```

Q5 Write a program to prepare a dictionary where the keys are numbers between 1 and 15 (both included) and the values are the squares of the keys.

```python
dict = {}

for i in range(1, 16):
    dict[i] = i * i

print(dict)
```

Q6 Write a program to sum all the values in a dictionary.

```python
dict = {
    1: 10,
    2: 20,
    3: 30,
    4: 40
}

total = sum(dict.values())

print("Sum of all values:", total)
```
