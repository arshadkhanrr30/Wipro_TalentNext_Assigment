Q1 Write a program to remove a given item from the set.

```python
set = {1, 2, 3, 4, 5}

item = int(input("Enter the item to remove"))

if item in set:
    set.remove(item)
    print("New Set:", set)
else:
    print("Item not found.")
```

Q2 Write a program to create an intersection of sets.

```python
set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

result = set1.intersection(set2)

print("Intersection is :", result)
```

Q3 Write a program to create a union of sets.

```python
set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

result = set1.union(set2)

print("Union is", result)
```

Q4 Write a program to find the maximum and minimum value in a set.

```python
set = {10, 50, 20, 40, 30 , 60 , 70 , 80 , 90}

maximum = max(set)
minimum = min(set)

print("Maximum value is:", maximum)
print("Minimum value is:", minimum)
```
