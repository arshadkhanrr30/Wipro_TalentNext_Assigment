Q1 Write a program to check if a given number is Positive, Negative, or Zero.

```python
num = float(input("Enter a number"))

if num > 0:
    print("The number is Positive")
elif num < 0:
    print("The number is Negative")
else:
    print("The number is Zero")
```

Q6 Write a program to check if a given number is prime or not.

```python
num = int(input("Enter a number"))

if num <= 1:
    print("Not Prime")
else:
    prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            prime = False
            break

    if prime:
        print("Prime")
    else:
        print("Not Prime")
```

Q7 Write a program to print prime numbers between 10 and 99.

```python
for num in range(10, 100):
    prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            prime = False
            break

    if prime:
        print(num,end="\t")
```

Q8 Write a program to print the sum of all the digits of a given number.
Example: 
I/P: 1234 
O/P: 10

```python
num = int(input("Enter a number"))
sum = 0

while num > 0:
    digit = num % 10
    sum += digit
    num //= 10

print(sum)
```

Q9 Write a program to reverse a given number and print.
Example: 1
I/P: 1234
O/P: 4321 

Example: 2
I/P: 1004
O/P: 4001

```python
num = int(input("Enter a number"))

rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("reverse is:", rev)
```

Q10 Write a program to find if the given number is palindrome or not
Example: 1
I/P: 110011
O/P: 110011 is a palindrome. 

Example: 2
I/P: 1234
O/P: 1234 is not a palindrome.

```python
num = int(input("Enter a number"))

original_num = num
rev_num = 0

while num > 0:
    digit = num % 10
    rev_num = rev_num * 10 + digit
    num //= 10

if original_num == rev_num:
    print(original_num, "is a palindrome.")
else:
    print(original_num, "is not a palindrome.")
```
