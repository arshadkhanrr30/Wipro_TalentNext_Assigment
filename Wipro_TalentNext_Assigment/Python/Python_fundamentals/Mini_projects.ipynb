```python
distance = float(input("How far would you like to travel in miles? "))

if distance < 3:
    print("I suggest Bicycle to your destination.")
elif distance < 300 and distance >= 3:
    print("I suggest Motor-Cycle to your destination.")
else:
    print("I suggest Super-Car to your destination.")
```

```python

cost_per_hour = 0.51
cost_per_day = cost_per_hour * 24
cost_per_week = cost_per_day * 7
cost_per_month = cost_per_day * 30

budget = 918
days = budget / cost_per_day

print("Cost to operate one server per day in dollar", cost_per_day)
print("Cost to operate one server per week in dollar", cost_per_week)
print("Cost to operate one server per month in dollar", cost_per_month)
print("Number of days one server can operate with $918:", days)
```
